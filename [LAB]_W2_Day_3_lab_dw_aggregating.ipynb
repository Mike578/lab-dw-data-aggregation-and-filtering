0. Import the dataset

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
```

```python
df.head()
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
# Filter the DataFrame based on the conditions
filtered_df = df[(df['total_claim_amount'] > 1000) & (df['response'] == 'Yes')]

# Get the number of customers
num_customers = len(filtered_df)

# Display the number of customers
print("Number of customers:", num_customers)
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# Filter the DataFrame for customers who responded "Yes" to the last marketing campaign
filtered_df = df[df['response'] == 'Yes']

# Group by policy type and gender, and calculate the average total_claim_amount
average_claim_amount = filtered_df.groupby(['policy_type', 'gender'])['total_claim_amount'].mean()

# Display the result
print(average_claim_amount)
```

- We can observe the average claim amount for each combination of policy type and gender. This information can be useful for understanding the distribution of claim amounts and identifying any patterns or trends.

- By comparing the average claim amounts across different policy types and genders, we can identify potential areas for further analysis or investigation. For instance, if certain policy types or gender groups consistently have higher claim amounts, it may indicate specific risk factors or customer behaviors that warrant closer attention.

- Additionally, this analysis can inform decision-making processes related to marketing strategies, product offerings, and customer segmentation, helping to optimize business outcomes and enhance customer satisfaction.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Group by state and count the number of customers in each state
# grouping the df by the unique values in the 'state' column and then calculates the size of each group
state_counts = df.groupby('state').size()

# Filter the results to include only states where there are more than 500 customers
states_with_more_than_500_customers = state_counts[state_counts > 500]

# Display the filtered results
print(states_with_more_than_500_customers)
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# Group by education level and gender, and calculate maximum, minimum, and median customer lifetime value
summary_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])

# Display the summary statistics
print(summary_stats)
```

The clv seems to be optimal for people in high-school or below in terms of education level

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
# Convert 'effective_to_date' column to datetime format
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

# Extract month from 'effective_to_date' column
df['month'] = df['effective_to_date'].dt.month_name()

# Pivot the DataFrame to create the table
pivot_table = df.pivot_table(index='state', columns='month', aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table)
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
# Calculate the total number of policies sold by state
state_policies_count = df.groupby('state').size().sort_values(ascending=False)

# Select the top 3 states with the highest number of policies sold
top_3_states = state_policies_count.head(3).index.tolist()

# Filter the original DataFrame to include only data for the top 3 states
df_top_3_states = df[df['state'].isin(top_3_states)]

# Pivot the filtered DataFrame to create the desired table
pivot_table_top_3_states = df_top_3_states.pivot_table(index='state', columns='month', aggfunc='size', fill_value=0)

# Display the new DataFrame
print(pivot_table_top_3_states)
```
